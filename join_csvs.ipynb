```python
import os
import pandas as pd
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
img1='20220516_113718.jpg'
img2='20220516_113208.jpg'
csv_dir = '/content/drive/My Drive/microplasticos/microplasticos_1152/csvs/'
```

```python
out_df=pd.DataFrame()
for f in os.listdir(csv_dir):
  if img2 in f:
    df = pd.read_csv(csv_dir+f)
    n=f.find('grid_')
    pcords=', '.join(f[(n+5):].strip('.jpg.csv').split('_'))#picture coordinates
    #print(pcords)
    df['grid_coords'] = pcords
    out_df=pd.concat([out_df, df], ignore_index=True)

out_df.head()
display(out_df)
```

```python
del out_df['Unnamed: 0']
```
