In order to compare the results of my work with YOLOv5 to the previous best results at the task of microplastics classification, it's time to do some transfer learnign with my data on the model described in the paper by  [Lorenzo-Navarro et al. (2021)](https://www.sciencedirect.com/science/article/pii/S0048969720362574)
<br>
In this notebook, we are cropping the particles from our images and separating them into corresponding folders in order to be able to train vgg16 model on these.

```python
import os
from os import walk, getcwd
import pandas as pd

from collections import namedtuple
from PIL import Image, ImageFont, ImageDraw, ImageEnhance


from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
in_path='/content/drive/My Drive/microplasticos/images/train_1152'
lbl_path='/content/drive/My Drive/microplasticos/labels/train_1152'
out_0='/content/drive/My Drive/microplasticos/images/train_vgg16/0'
out_1='/content/drive/My Drive/microplasticos/images/train_vgg16/1'
out_2='/content/drive/My Drive/microplasticos/images/train_vgg16/2'
out_3='/content/drive/My Drive/microplasticos/images/train_vgg16/3'
out_4='/content/drive/My Drive/microplasticos/images/train_vgg16/4'
```

```python
in_dir=os.listdir(in_path)
```

```python
for x in in_dir:
  print(lbl_path+'/'+x[:-4]+'.txt')
  lbl_dir=open(lbl_path+'/'+x[:-4]+'.txt', 'r')
  image=Image.open(os.path.join(in_path, x))
  lines = lbl_dir.read().split('\n')
  for line in lines:
    if line!='':
      line=line.split() #line: class, x, y, w, h
      print(line)
      c=line[0] #class
      left=(float(line[1])-0.5*float(line[3]))*1152 
      bot=(float(line[2])+0.5*float(line[4]))*1152
      top=bot-float(line[4])*1152
      right=left+float(line[3])*1152
      if bot>=1152:
        bot=1151.9
      if right>=1152:
        right=1151.9
      

      print(image.size)
      print(left, top, right, bot)
      crop_img = image.crop((left, top, right, bot))
      
      try:
        if c=='0':
          crop_img.save(os.path.join(out_0, x + "_tblr_" + str(round(top,2)) + "_" + str(round(bot,2)) + "_" + str(round(left,2))+"_" + str(round(right,2))+".jpg"))
        elif c=='1':
          crop_img.save(os.path.join(out_1,  x + "_tblr_" + str(round(top,2)) + "_" + str(round(bot,2)) + "_" + str(round(left,2))+"_" + str(round(right,2))+".jpg"))
        elif c=='2':
          crop_img.save(os.path.join(out_2, x + "_tblr_" + str(round(top,2)) + "_" + str(round(bot,2)) + "_" + str(round(left,2))+"_" + str(round(right,2))+".jpg"))
        elif c=='3':
          crop_img.save(os.path.join(out_3, x + "_tblr_" + str(round(top,2)) + "_" + str(round(bot,2)) + "_" + str(round(left,2))+"_" + str(round(right,2))+".jpg"))
        elif c=='4':
          crop_img.save(os.path.join(out_4,  x + "_tblr_" + str(round(top,2)) + "_" + str(round(bot,2)) + "_" + str(round(left,2))+"_" + str(round(right,2))+".jpg"))
      except:
        pass
        
```
